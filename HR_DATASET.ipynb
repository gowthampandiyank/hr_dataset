```python
import pandas as pd

# Load all sheets
data = pd.read_excel("HR_DATABASE.xlsx", sheet_name=None)

# Define age group labels
labels = ['Young', 'Adult', 'Senior']

# Loop through all sheets and create age_group where possible
for name, df in data.items():
    print(f"\nProcessing sheet: {name}")
    
    # Check if 'age' column exists in the sheet
    if 'age' in df.columns:
        # Create age_group using qcut
        df['age_group'] = pd.qcut(df['age'].rank(method='first'), q=3, labels=labels)
        
        # Show sample output
        print(df[['age', 'age_group']].head(10))
    else:
        print("No 'age' column found — skipping age_group creation.")
```
